    <h2 style="border: 0; border-radius: 15px; font-weight: bold; font-size: 220%; color: #FF5733; background-color: white; padding: 10px; text-align: center; text-shadow: 2px 2px 4px rgba(0, 0, 0, 0.5);">‚ö°Ô∏è Additive and Multiplicative Decomposition in Time Series Analysis ‚ö°Ô∏è</h2>
    <p>
        Time series analysis involves examining data that changes over time. To understand and model time series data, we often decompose it into its components. The main components are typically trend, seasonality, and randomness. Decomposition methods are used to analyze these components, and they come in two main types: <strong>additive</strong> and <strong>multiplicative</strong> decomposition.
    </p>
    <h3 style="border: 0; border-radius: 15px; font-weight: bold; font-size: 180%; color: #FF5733; background-color: white; padding: 10px; text-align: center; text-shadow: 2px 2px 4px rgba(0, 0, 0, 0.5);">Additive Decomposition</h3>
    <p>
        Additive decomposition breaks the time series into the following components:
    </p>
    <pre><code>Y_t = T_t + S_t + E_t</code></pre>
    <ul>
        <li><strong>Y_t</strong>: Observed value (time series)</li>
        <li><strong>T_t</strong>: Trend component</li>
        <li><strong>S_t</strong>: Seasonal component</li>
        <li><strong>E_t</strong>: Random component (error)</li>
    </ul>
    <p>
        Additive decomposition assumes that the seasonal component is added to the other components of the time series. This method implies that each component is independent and combined through addition.
    </p>
    <h4 style="border: 0; border-radius: 15px; font-weight: bold; font-size: 160%; color: #FF5733; background-color: white; padding: 10px; text-align: center; text-shadow: 2px 2px 4px rgba(0, 0, 0, 0.5);">Use Cases</h4>
    <ul>
        <li>Suitable when seasonal effects are pronounced in the data.</li>
        <li>Useful when seasonal effects do not change over time or have a constant cycle.</li>
    </ul>
    <h3 style="border: 0; border-radius: 15px; font-weight: bold; font-size: 180%; color: #FF5733; background-color: white; padding: 10px; text-align: center; text-shadow: 2px 2px 4px rgba(0, 0, 0, 0.5);">Multiplicative Decomposition</h3>
    <p>
        Multiplicative decomposition breaks the time series into the following components:
    </p>
    <pre><code>Y_t = T_t √ó S_t √ó E_t</code></pre>
    <ul>
        <li><strong>Y_t</strong>: Observed value (time series)</li>
        <li><strong>T_t</strong>: Trend component</li>
        <li><strong>S_t</strong>: Seasonal component</li>
        <li><strong>E_t</strong>: Random component (error)</li>
    </ul>
    <p>
        Multiplicative decomposition assumes that the seasonal component is multiplied with the other components of the time series. This method implies that seasonal effects vary with the magnitude of the time series and are combined through multiplication.
    </p>
    <h4 style="border: 0; border-radius: 15px; font-weight: bold; font-size: 160%; color: #FF5733; background-color: white; padding: 10px; text-align: center; text-shadow: 2px 2px 4px rgba(0, 0, 0, 0.5);">Use Cases</h4>
    <ul>
        <li>Suitable when seasonal effects vary with the size of the data.</li>
        <li>Useful when seasonal effects are related to the magnitude of the data and change over time.</li>
    </ul>
    <h3 style="border: 0; border-radius: 15px; font-weight: bold; font-size: 180%; color: #FF5733; background-color: white; padding: 10px; text-align: center; text-shadow: 2px 2px 4px rgba(0, 0, 0, 0.5);">Comparison and Selection</h3>
    <ul>
        <li><strong>Additive Decomposition</strong>: Generally suitable for data with fixed and distinct seasonal effects.</li>
        <li><strong>Multiplicative Decomposition</strong>: Suitable for data where seasonal effects change over time or are related to the size of the data.</li>
    </ul>
    <p>
        Using both methods to analyze time series data can help determine the most appropriate model based on the nature of the data.
    </p>
</div>

<a id="part_2_4"></a>
# <p style="background-color:#3498DB; font-family:newtimeroman;color:#FFF9ED; font-size:150%; text-align:center; border-radius: 15px 50px;"> ‚ú® Both Decomposition ‚ú®</p>


<a href="#part_2" class="btn btn-danger btn-lg active" role="button" aria-pressed="true" style="color:white;">Part - 2</a>
<a href="#toc" class="btn btn-warning btn-lg active" role="button" aria-pressed="true" style="color:black;">Content</a>

```python
decomposition_1 = sm(df["IPG2211A2N"], model = "add", period = 12)
decomposition_2 = sm(df["IPG2211A2N"], model = "mul", period = 12)
```

```python
decomposition_1.plot()
plt.title("Additive Decomposition")
decomposition_1.plot()
plt.title("Multiplicative Decomposition")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create figure with specified size
plt.figure(figsize=(12, 12))  # Width and height can be adjusted

# First subplot (top)
plt.subplot(3, 2, 1)  # 3 rows, 2 columns, 1st subplot
plt.plot(decomposition_1.trend, label='Trend')
plt.title('Additive Model - Trend')
plt.legend()

plt.subplot(3, 2, 3)  # 3 rows, 2 columns, 3rd subplot
plt.plot(decomposition_1.seasonal, label='Seasonal')
plt.title('Additive Model - Seasonal')
plt.legend()

plt.subplot(3, 2, 5)  # 3 rows, 2 columns, 5th subplot
plt.plot(decomposition_1.resid, label='Residual')
plt.title('Additive Model - Residual')
plt.legend()

# Second subplot (bottom)
plt.subplot(3, 2, 2)  # 3 rows, 2 columns, 2nd subplot
plt.plot(decomposition_2.trend, label='Trend')
plt.title('Multiplicative Model - Trend')
plt.legend()

plt.subplot(3, 2, 4)  # 3 rows, 2 columns, 4th subplot
plt.plot(decomposition_2.seasonal, label='Seasonal')
plt.title('Multiplicative Model - Seasonal')
plt.legend()

plt.subplot(3, 2, 6)  # 3 rows, 2 columns, 6th subplot
plt.plot(decomposition_2.resid, label='Residual')
plt.title('Multiplicative Model - Residual')
plt.legend()

# Display the plots
plt.tight_layout()  # Automatically adjusts layout for better spacing
plt.show()
```

<a href="#toc" class="btn btn-warning btn-lg active" role="button" aria-pressed="true" style="color:black;">Content</a>

<div style="background-color:#3498DB; padding:10px; border-radius: 10px; box-shadow: 2px 2px 4px rgba(0, 0, 0, 0.5);">
    <p style="font-weight: bold; color: white;">Looking at the residuals, it can be observed that the standard deviation of the additive model is higher than that of the multiplicative model. Consequently, if the additive model is used, its forecasting power will be somewhat lower compared to the multiplicative model.</p>
</div>


```python
# Calculate the standard deviation of the residuals for the Additive Model
std_residuals_additive = np.std(decomposition_1.resid)

# Print the result with a descriptive message
print(f"Standard Deviation of Residuals for the Additive Model: {std_residuals_additive:.4f}")
```

```python
# Calculate the standard deviation of the residuals for the Multiplicative Model
std_residuals_multiplicative = np.std(decomposition_2.resid)

# Print the result with a descriptive message
print(f"Standard Deviation of Residuals for the Multiplicative Model: {std_residuals_multiplicative:.4f}")
```

<a id="part_2_5"></a>
# <p style="background-color:#3498DB; font-family:newtimeroman;color:#FFF9ED; font-size:150%; text-align:center; border-radius: 15px 50px;"> ‚ú® Decomposition with Logarithm ‚ú®</p>


<a href="#part_2" class="btn btn-danger btn-lg active" role="button" aria-pressed="true" style="color:white;">Part - 2</a>
<a href="#toc" class="btn btn-warning btn-lg active" role="button" aria-pressed="true" style="color:black;">Content</a>

```python
decomposition_3 = sm(np.log(df["IPG2211A2N"]), model = "add", period = 12)
decomposition_4 = sm(np.log(df["IPG2211A2N"]), model = "mul", period = 12)
```

```python
# Create figure with specified size
plt.figure(figsize=(15, 15))  # Adjusted for additional plots

# First subplot (top-left)
plt.subplot(4, 4, 1)  # 4 rows, 4 columns, 1st subplot
plt.plot(decomposition_1.trend, label='Trend')
plt.title('Additive Model - Trend')
plt.legend()

# Second subplot (top-right)
plt.subplot(4, 4, 2)  # 4 rows, 4 columns, 2nd subplot
plt.plot(decomposition_2.trend, label='Trend')
plt.title('Multiplicative Model - Trend')
plt.legend()

# Third subplot (top-left)
plt.subplot(4, 4, 3)  # 4 rows, 4 columns, 3rd subplot
plt.plot(decomposition_3.trend, label='Trend')
plt.title('Log Additive Model - Trend')
plt.legend()

# Fourth subplot (top-right)
plt.subplot(4, 4, 4)  # 4 rows, 4 columns, 4th subplot
plt.plot(decomposition_4.trend, label='Trend')
plt.title('Log Multiplicative Model - Trend')
plt.legend()

# Fifth subplot (middle-left)
plt.subplot(4, 4, 5)  # 4 rows, 4 columns, 5th subplot
plt.plot(decomposition_1.seasonal, label='Seasonal')
plt.title('Additive Model - Seasonal')
plt.legend()

# Sixth subplot (middle-right)
plt.subplot(4, 4, 6)  # 4 rows, 4 columns, 6th subplot
plt.plot(decomposition_2.seasonal, label='Seasonal')
plt.title('Multiplicative Model - Seasonal')
plt.legend()

# Seventh subplot (middle-left)
plt.subplot(4, 4, 7)  # 4 rows, 4 columns, 7th subplot
plt.plot(decomposition_3.seasonal, label='Seasonal')
plt.title('Log Additive Model - Seasonal')
plt.legend()

# Eighth subplot (middle-right)
plt.subplot(4, 4, 8)  # 4 rows, 4 columns, 8th subplot
plt.plot(decomposition_4.seasonal, label='Seasonal')
plt.title('Log Multiplicative Model - Seasonal')
plt.legend()

# Ninth subplot (bottom-left)
plt.subplot(4, 4, 9)  # 4 rows, 4 columns, 9th subplot
plt.plot(decomposition_1.resid, label='Residual')
plt.title('Additive Model - Residual')
plt.legend()

# Tenth subplot (bottom-right)
plt.subplot(4, 4, 10)  # 4 rows, 4 columns, 10th subplot
plt.plot(decomposition_2.resid, label='Residual')
plt.title('Multiplicative Model - Residual')
plt.legend()

# Eleventh subplot (bottom-left)
plt.subplot(4, 4, 11)  # 4 rows, 4 columns, 11th subplot
plt.plot(decomposition_3.resid, label='Residual')
plt.title('Log Additive Model - Residual')
plt.legend()

# Twelfth subplot (bottom-right)
plt.subplot(4, 4, 12)  # 4 rows, 4 columns, 12th subplot
plt.plot(decomposition_4.resid, label='Residual')
plt.title('Log Multiplicative Model - Residual')
plt.legend()

# Display the plots
plt.tight_layout()  # Automatically adjusts layout for better spacing
plt.show()
```

```python
std_residuals_log_additive = np.std(decomposition_3.resid)

print(f"Standard Deviation of Residuals for the Log Additive Model: {std_residuals_log_additive:.4f}")
```

```python
std_residuals_log_multiplicative = np.std(decomposition_4.resid)

print(f"Standard Deviation of Residuals for the Log Multiplicative Model: {std_residuals_log_multiplicative:.4f}")
```
