## <p style="background-color:#9d4f8c; font-family:newtimeroman; color:#FFF9ED; font-size:150%; text-align:center; border-radius:10px 10px;">ABOUT THE DATASET & CONTEXT</p>

Welcome to **"Bike Demand Visualization Project"** which is the capstone project of Data Visualization Lessons . As you know recently, free or affordable access to bicycles has been provided for short-distance trips in an urban area as an alternative to motorized public transport or private vehicles. Thus, it is aimed to reduce traffic congestion, noise and air pollution.  

The aim of this project is to reveal the current patterns in the data by showing the historical data of London bike shares with visualization tools.

This will allow us to X-ray the data as part of the EDA process before setting up a machine learning model.

## <p style="background-color:#9d4f8c; font-family:newtimeroman; color:#FFF9ED; font-size:150%; text-align:center; border-radius:10px 10px;">ABOUT THE FEATURES</p>

<a id="00"></a>
<a href="#toc" class="btn btn-primary btn-sm" role="button" aria-pressed="true" 
style="color:blue; background-color:#dfa8e4" data-toggle="popover">Content</a>

**- ``timestamp``** - timestamp field for grouping the data<br>
**- ``cnt``** - the count of a new bike shares<br>
**- ``t1``** - real temperature in C<br>
**- ``t2``** - temperature in C “feels like”<br>
**- ``hum``** - humidity in percentage<br>
**- ``wind_speed``** - wind speed in km/h<br>
**- ``weather_code``** - category of the weather<br>
**- ``is_holiday``** - boolean field - 1 holiday / 0 non holiday<br>
**- ``is_weekend``** - boolean field - 1 if the day is weekend<br>
**- ``season``** - category field meteorological seasons: 0-spring ; 1-summer; 2-fall; 3-winter.<br>

**"weather_code" category description:**<br>
**``1``** = Clear ; mostly clear but have some values with haze/fog/patches of fog/ fog in vicinity<br> 
**``2``** = scattered clouds / few clouds<br> 
**``3``** = Broken clouds<br> 
**``4``** = Cloudy<br> 
**``7``** = Rain/ light Rain shower/ Light rain<br> 
**``10``** = rain with thunderstorm<br> 
**``26``** = snowfall<br> 
**``94``** = Freezing Fog<br>

```python
week_status = df["is_weekend"].value_counts()

plt.pie(labels = ["weekday", "weekend"],x = week_status.values , shadow = True, autopct="%1.1f%%");
```

<a id="4"></a>
<button type="button" class="btn btn-lg btn-success" data-toggle="popover" title="Popover title">Result</button>


<font color="#1f77b4">The number of weekdays is significantly higher than the number of weekends,</font>
<font color="#d62728">almost three times greater.</font>


```python
week_status = df["is_weekend"].value_counts()
ax = sns.barplot(x=week_status.index, y=week_status.values)

total = sum(week_status.values)
for i, v in enumerate(week_status.values):
    percentage = v / total * 100
    ax.annotate(f"{percentage:.1f}% - {v}", xy=(i, v), ha='center', va='bottom', fontsize=10, color='black')

plt.title("Weekend Status Bar Plot")
plt.show()
```

style="color:blue; background-color:#dfa8e4" data-toggle="popover">Content</a>

```python
weather_code = df["weather_code"].value_counts()

sns.barplot(x = weather_code.index, y = weather_code.values);
```

<a id="4"></a>
<button type="button" class="btn btn-lg btn-success" data-toggle="popover" title="Popover title">Result</button>


<font color="#1f77b4">We observe that the weather with code 1.0 (Clear; mostly clear but have some values with haze/fog/patches of fog/fog in vicinity) has the highest count,</font>
<font color="#d62728">while the weather with code 26.0 (snowfall) has a lower count.</font>


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define custom color palette
custom_palette = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728"]

# Set the seaborn style
sns.set(style="whitegrid")

# Filter data for each season
spring_data = df[df["season"] == "spring"]
summer_data = df[df["season"] == "summer"]
fall_data = df[df["season"] == "fall"]
winter_data = df[df["season"] == "winter"]

# Create a FacetGrid with custom palette
g = sns.catplot(data=df, x="weather_code", col="season", kind="count", palette=custom_palette)

# Set axis labels and titles
g.set_axis_labels("Weather Code", "Count")
g.set_titles("Season = {col_name}")

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

<a id="4"></a>
<button type="button" class="btn btn-lg btn-success" data-toggle="popover" title="Popover title">Result</button>


<font color="#1f77b4">We observe that the weather code 26.0 (snowfall) is more frequent during the 3.0 season (winter).</font>
<br>
<font color="#ff7f0e">During the 1.0 season (summer), the weather code 2.0 (scattered clouds / few clouds) is more prevalent compared to other seasons,</font>
<font color="#d62728">while the weather code 3.0 (Broken clouds) is more frequent than code 2.0.</font>


```python
weather_season_data = df.groupby(["weather_code", "season"]).size().reset_index(name="count")

ax = sns.barplot(x="weather_code", y="count", hue="season", data=weather_season_data)
ax.legend(loc = "upper center", bbox_to_anchor = (1.15, 0.5))
plt.show()
```
